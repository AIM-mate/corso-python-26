### Soluzione:


````python
raw = [" 10", 7, "3.5", None, " -2 ", "errore", 4.0, "0", "", "  8  ", "+12", "-0.75", "1.2.3", "abc123", "  -0  ", -9, 2.0]

# TODO: costruisci values

values = []
for x in raw:
	
	# Ignora valori mancanti
	if x is None:
		continue

	# Caso 1: valore stringa da ripulire e validare
	if isinstance(x, str):
		s = x.strip()
		if s == "":
			continue

		# Rimuove il segno solo per la validazione
		core = s.lstrip("+-")
		if core == "":
			continue
		if core.count(".") > 1:
			continue
		if not core.replace(".", "", 1).isdigit():
			continue

		# Converte mantenendo il segno originale (usa s, non core)
		num = float(s) if "." in core else int(s)
		values.append(num)

	# Caso 2: valore già numerico
	elif isinstance(x, (int, float)):
		if isinstance(x, float) and x.is_integer():
			values.append(int(x))
		else:
			values.append(x)

print(f"Valori puliti: {values}\n")

# TODO: calcola statistiche e stampa report

report = (
	f"REPORT:\n"
	f"Min: {min(values)}\n"
	f"Max: {max(values)}\n"
	f"Somma: {sum(values)}\n"
	f"Media: {sum(values) / len(values):.3f}\n"
	f"Negative: {sum(1 for v in values if v < 0)}")

print(report)
````

cart = ["pane", "latte", "PANE", "uova", "VINO", "latte", "", None, "  pasta  ", "uova"]

E un listino prezzi:

prices = {"pane": 1.20, "latte": 1.50, "uova": 0.30, "pasta": 0.90}

4. Stampa una “fattura” ordinata alfabeticamente per prodotto, con subtotale e totale.

---

### Spazio di lavoro:


````python
cart = ["pane", "latte", "PANE", "uova", "VINO", "latte", "", None, "  pasta  ", "uova"]
prices = {"pane": 1.20, "latte": 1.50, "uova": 0.30, "pasta": 0.90}

# TODO: normalizza cart -> clean
clean = []

# TODO: costruisci qty
qty = {}

# TODO: missing e totale
missing = None
total = None

print("")

# TODO: stampa fattura
print("FATTURA")
````

### Soluzione:


````python
cart = ["pane", "latte", "PANE", "uova", "VINO", "latte", "", None, "  pasta  ", "uova"]
prices = {"pane": 1.20, "latte": 1.50, "uova": 0.30, "pasta": 0.90}

# TODO: normalizza cart -> clean
clean = []
for item in cart:
	if item is None:
		continue
	if not isinstance(item, str):
		continue
	s = item.strip().lower()
	if s == "":
		continue
	clean.append(s)

# TODO: costruisci qty
qty = {}
for item in clean:
	qty[item] = qty.get(item, 0) + 1

# TODO: missing e totale
products_in_cart = set(qty)
products_in_prices = set(prices)
missing = products_in_cart - products_in_prices
total = sum([qty[p] * prices[p] for p in products_in_cart if p in prices])
print(f"Missing: {sorted(missing)}")
print(f"Total: {total:.2f}")
print("")

# TODO: stampa fattura
print("FATTURA")
print("-" * 40)
for product in sorted(qty):
	if product in prices:
		unit = prices[product]
		subtotal = qty[product] * unit
		print(f"{product:7s} x{qty[product]:2d} @ {unit:4.2f} = {subtotal:4.2f}")

print("-" * 40)
print(f"TOTALE: {total:.2f}")
````

## Esercizio 4

### Consegna:
Hai due liste:

```python
names  = ["Ada", "Bruno", "Ciro", "Dina", "Elio", "Fede", "Gina", "Hugo", "Iris", "Luca", "Mara", "Nico", "Omar", "Pina", "Rita", "Sami", "Tina", "Ugo", "Vera", "Zeno"]
scores = [28, 15, 30, 22, None, 18, 24, None, 27, 29, 30, 16, None, 21, 19, 25, None, 30, 14, 23]
```

Regole di classificazione:
- Se manca un voto, usa `None`.
- Esito:
	- `None` → `"ASSENTE"`
	- `< 18` → `"BOCCIATO"`
	- `18..24` → `"OK"`
	- `25..29` → `"BUONO"`
	- `30` → `"ECCELLENTE"`

Obiettivi:
1. Costruisci una lista di record `records` nella forma `(nome, voto, esito)`.
2. Costruisci un dizionario `by_name` che mappa ogni nome al suo esito.
3. Calcola:
	 - media dei soli voti presenti (ignora i `None`)
	 - nome/i con voto massimo (se ex-aequo, tienili tutti)
4. Stampa un report leggibile.

---

### Spazio di lavoro:

````python
names  = ["Ada", "Bruno", "Ciro", "Dina", "Elio", "Fede", "Gina", "Hugo", "Iris", "Luca", "Mara", "Nico", "Omar", "Pina", "Rita", "Sami", "Tina", "Ugo", "Vera", "Zeno"]
scores = [28, 15, 30, 22, None, 18, 24, None, 27, 29, 30, 16, None, 21, 19, 25, None, 30, 14, 23]

# TODO: crea records = [(name, score, status), ...] con classificazione inline
records = []

# TODO: crea by_name con dict comprehension
by_name = {}

# TODO: calcola average sui soli voti presenti
average = None

# TODO: trova max_score e top_students (gestendo eventuali ex-aequo)
max_score = None
top_students = []

# TODO: stampa report finale
print("RISULTATI")
````

### Soluzione:


````python
names  = ["Ada", "Bruno", "Ciro", "Dina", "Elio", "Fede", "Gina", "Hugo", "Iris", "Luca", "Mara", "Nico", "Omar", "Pina", "Rita", "Sami", "Tina", "Ugo", "Vera", "Zeno"]
scores = [28, 15, 30, 22, None, 18, 24, None, 27, 29, 30, 16, None, 21, 19, 25, None, 30, 14, 23]

# TODO: crea records = [(name, score, status), ...] con classificazione inline
records = []
for name, score in zip(names, scores):
	if score is None:
		status = "ASSENTE"
	elif score < 18:
		status = "BOCCIATO"
	elif score <= 24:
		status = "OK"
	elif score <= 29:
		status = "BUONO"
	else:
		status = "ECCELLENTE"
	records.append((name, score, status))

# TODO: crea by_name con dict comprehension
by_name = {name: status for name, _, status in records}

# TODO: calcola average sui soli voti presenti
present_scores = [score for _, score, _ in records if score is not None]
average = sum(present_scores) / len(present_scores)

# TODO: trova max_score e top_students (gestendo eventuali ex-aequo)
max_score = max(present_scores)
top_students = [name for name, score, _ in records if score == max_score]

# TODO: stampa report finale
print("RISULTATI")
print("-" * 40)
for name, score, status in records:
	score_text = "--" if score is None else str(score)
	print(f"{name:6s} score ={score_text:>3s} -> {status}")

print("-" * 40)
print(f"Media (presenti): {average:.2f}")
print(f"Massimo: {max_score} | Migliori: {top_students}")
print(f"Esiti: {by_name}")
````
