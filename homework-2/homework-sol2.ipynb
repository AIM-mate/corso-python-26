### Soluzione:


````python
def merge_configs(*configs, override=True):
	result = {}
	for cfg in configs:
		if not isinstance(cfg, dict):
			raise TypeError(f"Expected dict, got {type(cfg).__name__}")

		if not override:
			overlap = result.keys() & cfg.keys()
			if overlap:
				# set(...) per rendere il messaggio stabile/leggibile
				raise ValueError(f"Duplicate keys: {set(overlap)}")

		# unpacking: crea un nuovo dict senza modificare gli input
		result = {**result, **cfg}

	return result


a = {"x": 1, "y": 2}
b = {"y": 10, "z": 3}

print(merge_configs(a, b))  # {'x': 1, 'y': 10, 'z': 3}
print(a, b)                 # verifica: invariati

try:
	merge_configs(a, b, override=False)
except ValueError as e:
	print("Expected error:", e)
````

````python
def read_rows(path):
    with open(path, "r") as f:
        next(f, None)  # scarta header
        for line in f:
            line = line.strip()
            if line:
                yield line

def parse_row(line):
    id_s, age_s, income_s, city = line.split(",")
    return {"id": int(id_s), "age": int(age_s), "income": int(income_s), "city": city}

path = "ex2.2_mock_data.csv"
rows = read_rows(path)

data = map(parse_row, rows)
data = filter(lambda r: r["income"] != 0, data)
data = map(lambda r: {**r, "income_k": r["income"] / 1000}, data)

for item in data:
	print(item)
````

## Esercizio 4

### Consegna:
Devi implementare un **decoratore parametrico**:

```python
chunk_processor(size, *, strict=False)
```

che trasformi una funzione del tipo:

```python
f(chunk) -> valore
```

in una funzione decorata del tipo:

```python
f(iterable, *args, **kwargs) -> lista_di_valori
```

### Specifiche richieste:
1. Implementa la factory:
   - `def chunk_processor(size, *, strict=False): ...`
2. Valida i parametri:
   - se `size` non è un `int` oppure `size <= 0`, solleva `ValueError`.
3. Il decoratore deve restituire un `wrapper(iterable, *args, **kwargs)`.
4. Nel wrapper:
   - crea l’iteratore con `it = iter(iterable)`
   - consuma i dati usando **solo** `next(it)` con `try/except StopIteration`
   - non usare `itertools` (o moduli simili)
   - non convertire tutto `iterable` in lista
5. Costruisci chunk di lunghezza `size` e passali a `func` come `tuple`.
6. Comportamento di `strict`:
   - `strict=False`: se l’ultimo chunk è incompleto ma non vuoto, va processato comunque
   - `strict=True`: se l’ultimo chunk è incompleto, solleva `ValueError`
7. Valore di ritorno:
   - il wrapper restituisce una lista `results` con un elemento per ogni chunk elaborato
   - ogni chiamata deve essere: `func(chunk, *args, **kwargs)`

### Quick check atteso:
- Con `@chunk_processor(3)` e `sum_chunk(range(7))` ottieni: `[3, 12, 6]`
- Con `@chunk_processor(3, strict=True)` e `range(7)` viene sollevato `ValueError`

---

### Soluzione:


````python
def chunk_processor(size, *, strict=False):
	if not isinstance(size, int) or size <= 0:
		raise ValueError("size must be a positive int")

	def decorator(func):
		def wrapper(iterable, *args, **kwargs):
			it = iter(iterable)
			results = []

			while True:
				chunk = []
				for _ in range(size):
					try:
						chunk.append(next(it))
					except StopIteration:
						if not chunk:
							return results
						if strict:
							raise ValueError(
								f"incomplete chunk of length {len(chunk)} (expected {size})"
							)
						results.append(func(tuple(chunk), *args, **kwargs))
						return results

				results.append(func(tuple(chunk), *args, **kwargs))

		return wrapper
	return decorator


# quick check
@chunk_processor(3)
def sum_chunk(chunk):
	return sum(chunk)

print(sum_chunk(range(7)))

@chunk_processor(3, strict=True)
def sum_chunk_strict(chunk):
	return sum(chunk)

try:
	print(sum_chunk_strict(range(7)))
except ValueError as e:
	print("strict error:", e)
````
