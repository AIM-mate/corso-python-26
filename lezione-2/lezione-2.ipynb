### Ricorsione

La ricorsione è una tecnica di implementazione di funzioni tramite la chiamata a se stessi.
In alcuni casi (come il fattoriale) il risultato è estremamente leggibile e coinciso.

La ricorsione è quasi sempre l'opzione più lenta dal punto di vista del costo computazionale. Inoltre, una funzione non può chiamare se stessa all'infinito (in profondità).

```python
def recursive_factorial(n):
    if n == 0:
        return 1
    else:
        return n * recursive_factorial(n - 1)

result = recursive_factorial(10)
print(result)
```

```python
# dovrebbe dare errore "RecursionError: maximum recursion depth exceeded" se troppo grande
recursive_factorial(3000)
```

### Scope delle variabili

Introduciamo ora il concetto di **scope**. Lo scope è l'intervallo di durata di vita di una variabile.

```python
def local_variable_scope():
    local_variable = 10
    local_variable += 1
    print(local_variable)
local_variable_scope()
print(local_variable)
```

Come possiamo notare una variabile locale viene "creata" e "distrutta" all'interno del corpo della funzione, provandovi ad accedere fuori dal corpo della funzione otteniamo un errore

```python
global_variable = 10
def global_variable_write():
    global_variable += 1
    print(global_variable)
global_variable_write()
```

Mentre se abbiamo definito una variabile globale non ci risulta poi possibile modificarla all'interno del corpo della funzione. Per modificarla all'interno della funzione possiamo servirci del comando **global**

```python
global_variable = 10

def global_variable_write_with_global():
    global global_variable
    global_variable += 1
    print(global_variable)
global_variable_write_with_global()
```

## Funzioni anonime

Utilizzate per funzioni piccole. Utilizziamo la parola chiave **lambda** e la seguente sintassi

```python
#area cerchio (1 param in input)
circle_area = lambda x: x * x * 3.14
print(circle_area(2))

#somma (2 param in input)
mysum = lambda x, y: x + y
print(mysum(3,4))

# un caso d'uso perfetto
points = [(1, 2), (4, 1), (5, -3)]
# ordiniamo i punti per y crescenti
points.sort(key=lambda x: x[1])
print(points)
```

#### Quando utizzarle?
Le lambda sono come i post-it della programmazione: utili per scriverci sopra una nota al volo e buttarli via subito dopo. Se inizi a incollarli ovunque, stai creando un disastro. Ecco alcune indicazioni per capire se hanno senso
- Usa una lambda solo se la funzione ti serve una volta sola
- Deve essere leggibile in letteralmente 2 secondi
- Se serve un commento per spiegarla è sbagliata

## Gestione di Argomenti Variabili:`*args` e `**kwargs`
In Python, puoi definire funzioni che accettano un numero arbitrario di argomenti. Questo è utile per creare wrapper o gestire multiple configurazioni. 
Questa è una particolare applicazione dell'**unpacking** che abbiamo visto la scorsa volta.

```python
def sum_all(*args):
    # args è una tupla che contiene tutti i valori passati
    print(f"Argomenti ricevuti: {args} | Tipo: {type(args)}")
    return sum(args)

# Esempi di chiamata
print(sum_all(1, 2))          # Output: 3
print(sum_all(10, 20, 30, 5)) # Output: 65
```

```python
def describe_user(**kwargs):
    # kwargs è un dizionario
    for key, value in kwargs.items():
        print(f"{key}: {value}")

# Esempio di chiamata
describe_user(name="Alice", course="Informatica", year=2)
```

Naturalmete si possono mischiare, ma bisogna seguire un ordine rigoroso:
-    Argomenti standard (obbligatori).
-    `*args`
-    `**kwargs`

```python
def mixed_func(classic_arg, *args, **kwargs):
    print(f"Obbligatorio: {classic_arg}")
    print(f"Extra posizionali (tuple): {args}")
    print(f"Extra nominali (dict): {kwargs}")   

mixed_func("Test", 1, 2, 3, chiave="valore")
```

## Funzioni annidate e decorators

In Python, le funzioni sono oggetti come gli altri: possono essere passate come argomenti, restituite da altre funzioni e definite l'una dentro l'altra.

```python
def compute_tax(tax):
    def tax_fn(price):
        # leggiamo tax dall'outer scope
        return price * (1 + tax / 100.0)
    
    # stiamo restituendo una funzione
    return tax_fn

italian_tax = compute_tax(22)
print(italian_tax(10))
```

### Decorators
Un decoratore è semplicemente una funzione che prende un'altra funzione come input, ne estende il comportamento e la restituisce "impacchettata".

```python
import time

# creiamo una funzione the data una funzione 
# restituisce una funzione che la esegue e calcola il tempo di esecuzione
def how_long(fn_to_decorate):
    def wrapper():
        t0 = time.time()
        fn_to_decorate()
        print(f"--- Esecuzione: {time.time() - t0} ---")
    return wrapper

# la funzione da "decorare"
def say_hello():
    print("Hello!")

# la funzione "decorata"
say_hello_decorated = how_long(say_hello)

# eseguiamola!
say_hello_decorated()
```

Python è molto pragmatico: il "decorator" è molto utilizzato, vogliamo usarlo più semplicemente, quindi c'è una sintassi per fare questo più brevemente.

```python
# applichiamo "how_long" con "@"
@how_long
def say_hello_decorated_atsign():
    print("Hello!")

say_hello_decorated_atsign()
```

Ultimo step, cosa succede se alla nostra funzione sevono degli argomenti? Usiamo `**args` e `**kwargs`!

```python
def how_long_with_forward(fn_to_decorate):
    def wrapper(*args, **kwargs):
        t0 = time.time()
        fn_to_decorate(*args, **kwargs)
        print(f"--- Esecuzione: {time.time() - t0} ---")
    return wrapper

@how_long_with_forward
def say_hello_to_someone(name):
    print(f"Hello {name}!")

say_hello_to_someone("World")
```

## Error Management

Gli errori sono una parte naturale della programmazione. 
Python fornisce il meccanismo `try-except` per gestire gli errori.


### Opinioni forti
Fino a che non diventate programmatori avanzati **non** utilizzate questo meccanismo. Non si può non parlarne, ma abusare di questo meccanismo porta a codici illeggibili ed immantenibili. Questo meccanismo è molto attraente e troppo facilmente abusabile se non si è esperti.

```python
l = [1,2,3,4,5]
try:
    # questo elemento non esiste
    result = l[6]
    # questa line non viene eseguita
    print("Abbiamo preso l'elemento")
except:
    print("C'e' stato un errore ma continuiamo come se niente fosse")

# ATTENZIONE: result non esiste!
print("Fine")
```

In questo caso ignoriamo ogni errore. La scelta leggemente meno scellerata è quella di ignorare solo alcuni tipi di errore. Per fare ciò si specifica dopo `except` il tipo di eccezione che vogliamo ignorare.

```python
try:
    result = 10 / 0
except ZeroDivisionError as err:
    print("Exception:", err)

print("Fine")
```

Non voglio darvi altre informazioni poichè preferirei davvero non le utilizzaste. Ultima informazione, al contrario potete anche lanciare voi un errore per fermare l'esecuzione. Vi consiglio invece di usare questa tecnica per essere sicuri di aver gestito tutte le casistiche.

```python
def animal_sound(animal):
    if animal == "cat":
        print("miao")
    elif animal == "dog":
        print("wof")
    else:
        raise Exception(f"We do not have this animal '{animal}'")

animal_sound("dog")
animal_sound("bird")
```

## Lettura e scrittura di files

Leggere e scrivere files è davvero una delle parti chiave di ogni linguaggio di programmazione perchè permette di interagire con il sistema operativo e fare davvero cose interessanti.

Esistono diversi modi per leggere e scrive files, tuttavia, vi voglio mostrare l'unico modo che per me ha senso.

```python
# scriviamo un file che ha questo nome
filename = "./my-file.txt"

# nuova sintassi:
# - la "w" sta per write, stiamo chiedendo al'OS di scrivere un file (se esiste viene SOVRASCRITTO!)
#   per aggiungere ad un file esistente usare la modalita' "a" (append)
# - "fp" è il nome della variabile che contiene il file, va bene anche se ha 2 lettere
with open(filename, "w") as fp:
    fp.write("Hello from this file!")

# appena cambiamo indentazione il file viene chiuso automaticamente 
```

Ora leggiamo il nostro file

```python
with open(filename, "r") as fp:
    content = fp.read()

print(content)
```

Esistono opzioni per chiudere manualmente il file e per scrivere o leggere riga per riga. Queste opzioni portano generalmente a codice con errori o costoso. Chiedere all'OS le risorse del file system dovrebbe essere fatto con cura.

La maggior parte degli utilizzi possono essere risolti con metodi per stringhe (es. `"\n".join(...)` per unire righe e `content.split("\n")` per dividerle).

#### JSON
Il formato JSON è ottimo per dati non strutturati, inoltre è altamente utilizzato. Vi basti pensare che molta della communicazione che avviene su internet avviene tramite JSON. Quindi molti linguaggi supportano la lettura e scrittura di JSON.

Il JSON è un formato plaintext, significa che occupa più spazio ma è completamente leggibile da umani.

Un modo semplice per pensare ad un JSON per noi è come la traduzione 1-a-1 (in entrambi i sensi) di qualuque combinazione delle seguenti strutture Python (innestate a piacere):
- Dizionari `dict`
- Liste `list`
- Interi `int`
- Numeri con la virgola `float`
- Stringhe `str`
- Booleani `bool`
- `None`

```python
# oggetto complicato legato ad un profilo utente, i dettagli non ci interessano
user_profile = {
    "id": 101,
    "account_info": {
        "username": "coder_pro",
        "email": "dev@example.com",
        "settings": {
            "theme": "dark",
            "notifications": False
        }
    },
    "interests": ["python", "api", "json"],
    "history": [
        {"action": "login", "timestamp": "2026-02-12"},
        {"action": "upload", "timestamp": "2026-02-13"}
    ]
}

import json

# salviamo l'oggetto molto facilmente con una sintassi simile a quella di scrittura file
with open("user_profile.json", "w") as fp:
    json.dump(user_profile, fp)

# lo leggiamo altrettanto facilmente con una sintassi quasi uguale
with open("user_profile.json", "r") as fp:
    user_profile_read = json.load(fp)

# controlliamo che i due oggetti siano uguali
user_profile == user_profile_read
```

Rendiamo le cose più pepate, facciamo finta di avere dei dati che arrivano da delle stazioni meteo. Per prima cosa facciamo finta di crearli.

```python
import random

# creiamo una cartella per le nostre misurazioni
os.makedirs("./weather", exist_ok=True)

for day in range(31):
    # facciamo finta di creare una misurazione
    measurement = {
        "day": f"2026-02-{day + 1:02}",
        "temperature_C": random.random() * 10,
        "rain_mm": random.random() * 5,
    }

    # la salviamo
    with open(f"./weather/{measurement['day']}.json", "w") as fp:
        json.dump(measurement, fp)
```

Questo formato (che è molto simile a quello che potreste incontrare nella vita reale quando qualcuno vi da dei dati), è decisamente scomodo per due motivi:
1. Divisi in multipli files
2. Non sfruttiamo il fatto che i dati sono strutturati (ogni giorno abbiamo gli stessi campi)

Tuttavia viene in nostro aiuto il formato 

### CSV (Comma Separated Values)

Pensate ad un CSV come una tabella, praticamente un Excel senza colori o altre amenità.
La libreria per gestire CSV è `pandas`. In realtà, essa è molto di più (sarebbe come dire che la Ferrari fa la vernice rossa), ma andiamo con calma.

```python
import pandas

# controlliamo tutti i file json che abbiamo nella cartella 
# cosi` se ne abbiamo di nuovi ogni giorno il codice funziona lo stesso
weather_files = os.listdir("./weather")
weather_files = [file for file in weather_files if file.endswith(".json")]
# ordine cronologico
weather_files.sort()

# mettiamoli tutti insieme in una lista
weather_measures = []
for file in weather_files:
    with open(f"./weather/{file}", "r") as fp:
        weather_measures.append(json.load(fp))

# ecco pandas che fa la sua magia, prende dati non strutturati e li trasforma in tabella
weather_dataframe = pandas.DataFrame(weather_measures)
print(weather_dataframe)

# salviamo la tabella in formato CSV
# index=False vuol dire che non salviamo il numero di riga
weather_dataframe.to_csv("./weather/all_measurements.csv", index=False)
```

### Per tutto il resto c'è ~~Mastercard~~ `pickle`

Le opzioni che vi ho mostrato fino ad ora possono sembrarvi un po' limitanti per il tipo di dati che si possono salvare/leggere. Python ha un arma potentissima nel suo arsenale `pickle`. Come molte delle armi potenti che vi ho mostrato, ve lo **SCONSIGLIO**. In parte, perchè promuove cattivi pattern di programmazione e in parte poichè vi espone a rischi di sicurezza (qualcuno vi può passare un file `pickle` malevolo che funziona esattamente come un virus appena caricato su Python, rubandovi o cancellandovi dati sensibili).

`pickle` è un formato binario, quindi non leggibile dall'uomo e molto leggero. Inoltre, rende possibile salvare praticamente qualunque oggetto Python poichè salva la sua rappresentazione in memoria (RAM).

Ecco un mini tutorial

```python
import pickle

data = {
    "ids": {1, 2, 3},  # Sets cannot be saved in JSON!
    "meta": "Internal state"
}

# Saving (Pickling), notate "wb" == write+binary!!!
with open("data.pkl", "wb") as f:
    pickle.dump(data, f)

# Loading (Unpickling), notate "rb" == write+binary!!!
with open("data.pkl", "rb") as f:
    restored_data = pickle.load(f)

data == restored_data
```
